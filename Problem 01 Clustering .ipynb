```python
def remove_outlier(col):
    sorted(col)
    Q1, Q3 = col.quantile([0.25, 0.75])
    IQR = Q3-Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range =  Q3 + (1.5 * IQR)
    return lower_range, upper_range 
```

```python
lratio,uratio=remove_outlier(bk['probability_of_full_payment'])
bk['probability_of_full_payment']=np.where(bk['probability_of_full_payment']>uratio,uratio,bk['probability_of_full_payment'])
bk['probability_of_full_payment']=np.where(bk['probability_of_full_payment']<lratio,lratio,bk['probability_of_full_payment'])

lraxis,uraxis=remove_outlier(bk['min_payment_amt'])
bk['min_payment_amt']=np.where(bk['min_payment_amt']>uraxis,uraxis,bk['min_payment_amt'])
bk['min_payment_amt']=np.where(bk['min_payment_amt']<lraxis,lraxis,bk['min_payment_amt'])
```

```python
from scipy.stats import zscore
bk_new=bk.apply(zscore)
bk_new.head()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
wardlink = linkage(bk_new, method = 'ward')
```

```python
plt.figure(figsize = (25,10))
plt.title("DENDOGRAM")
dend = dendrogram(wardlink)
```

```python
plt.figure(figsize = (25,10))
plt.title("DENDOGRAM")
dend = dendrogram(wardlink)
plt.axhline(y=25 , color = 'y' , linestyle = '--')
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
clusters = fcluster(wardlink, 25, criterion='distance')
clusters
```

```python
bk_hier = bk.copy()
```

```python
bk_hier['clusters'] = clusters
```
